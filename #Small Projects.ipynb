```python
c=0
c1=0
while True:
    if c==5:
        print("YOU WINS THE GAME...😊")
        break
    elif c==5:
        print("COMPUTER WIN THE GAME...🤷‍♀️")
        break
    import random
    l=["stone","paper","scissor"]
    random.shuffle(l)
    m=input("enter your action :1.STONE 2.PAPER 3.SCISSOR      :")
    print("comuter selected :",l[0])
    if l[0]=="stone" and m=="1":
        print("draw")
    elif l[0]=="stone" and m=="2":
        c=c+1
        print("you wins",c)
    elif l[0]=="stone" and m=="3":
        c1=c1+1
        print("computer wins",c1)
    elif l[0]=="paper" and m=="1":
        c1=c1+1
        print("computer wins",c1)
    elif l[0]=="paper" and m=="2":
        print("draw")
    elif l[0]=="paper" and m=="3":
        c=c+1
        print("you wins",c)
    elif l[0]=="scissor" and m=="1":
        c=c+1
        print("you wins",c)
    elif l[0]=="scissor" and m=="2":
        c1=c1+1
        print("computer wins",c1)
    elif l[0]=="scissor" and m=="3":
        print("draw")
```
